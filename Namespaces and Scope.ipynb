## Namespaces:



### Introduction to Names and Namespaces

- A name (symbolic name) is an ideitifier for an object. In the below code, the name of color is assigned to the string cyan

```Python
color = "cyan"
```

- A namespace is a collection of names and the objects that they reference.The namespace is a dictionary with the keys are the names and mapped values are the referenced objects.
- The namespace for the above code would be {'color': 'cyan'}

There are 4 distinct types of namespaces:
- Built-in
- Global
- Local
- Enclosing

### Built-in Namespace

- print() and str() are available all the time as their names of print and str are in the built-in namespace
- Built-in namespace is at the highest level of namespaces, thus can be called in any program we write
- dir(\_\_builtins\_\_) gives the dictionary with names as keys in the built-in namespace

````python
print(dir(__builtins__))
print("---")
print(f"There are {len(dir(__builtins__))} names in the built-in Namespaces and Scope.ipynb")
````

````python
__builtins__
````

From the above list, we can see that built-in namespaces include:
- Built-in functions such as str(), type()
- Built-in exception such as ArithmeticError, AssertionError
- Built-in constants such as True, False

### Global Namespace

- The global namespace is one level below the built-in namespace.
- It generally includes all non-nested names in teh module (file) we are running Python interpreter on.
- globals() give the dictionary of the global namespace. globals is the name in the built-in namespace.

````python
#Imaginary File: main.py
import random
 
first_name = "Jaya"
last_name = "Bodegard" 
 
def print_variables():
    random_number = random.randint(0,9)
    print(first_name)
    print(last_name)
    print(random_number)

print(globals())  
````

### Local Namespace
- It is the deepest level

````python
global_variable = 'global'
 
def add(num1, num2):
  nested_value = 'Inside Function'   
  print(num1 + num2)
  
  # This is to print the locals of the function add 
  print(locals())
 
add(5, 10)

# This is to print the local of the outside of the add function, which is
# actually global level. Thus global was printout.
print(locals())
````

### Enclosing namespace:
A special type of local namespace that occurs when workign with nested function.

````python
global_variable = 'global'
 
def outer_function():
  outer_value = "outer"
 
  def inner_function():
    inner_value = "inner"
  inner_function()
  # Added locals output
  print(locals())
 
outer_function()
````

## Scopes

Generally, inside scope can refer variable of outside scope but cannot change the value.

Below is teh example, which teh nested_function try to change the var of the outer. However, the value does not change when printing var in the outer level.

````python
def enclosing_function():
  var = "value"
 
  def nested_function():
    var = "new_value"
    print(f"var in nested function is: {var}")
    print(f"var id is: {id(var)}")
 
  nested_function()
 
  print(f"var in the enclosing_function is: {var}")
  print(f"var id is: {id(var)}")
 
enclosing_function()
````

In order to change the outer variable, we need to use **nonlocal** before changing its value.

````python
def enclosing_function():
  var = "value"
 
  def nested_function():
    nonlocal var
    var = "new_value"
 
  nested_function()
  print(var)
 
enclosing_function()
````

Similarly, we can use keyword **global** to modify the global variable.

````python
global_var = 10
 
def some_function():
  global global_var
  global_var = 20
 
some_function()
 
print(global_var)
````
