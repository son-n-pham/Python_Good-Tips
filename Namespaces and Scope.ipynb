## Namespaces:



### Introduction to Names and Namespaces

- A name (symbolic name) is an ideitifier for an object. In the below code, the name of color is assigned to the string cyan

```Python
color = "cyan"
```

- A namespace is a collection of names and the objects that they reference.The namespace is a dictionary with the keys are the names and mapped values are the referenced objects.
- The namespace for the above code would be {'color': 'cyan'}

There are 4 distinct types of namespaces:
- Built-in
- Global
- Local
- Enclosing

### Built-in Namespace

- print() and str() are available all the time as their names of print and str are in the built-in namespace
- Built-in namespace is at the highest level of namespaces, thus can be called in any program we write
- dir(\_\_builtins\_\_) gives the dictionary with names as keys in the built-in namespace

````python
print(dir(__builtins__))
print("---")
print(f"There are {len(dir(__builtins__))} names in the built-in Namespaces and Scope.ipynb")
````

From the above list, we can see that built-in namespaces include:
- Built-in functions such as str(), type()
- Built-in exception such as ArithmeticError, AssertionError
- Built-in constants such as True, False

### Global Namespace

- The global namespace is one level below the built-in namespace.
- It generally includes all non-nested names in teh module (file) we are running Python interpreter on.
- globals() give the dictionary of the global namespace. globals is the name in the built-in namespace.

````python
#Imaginary File: main.py
import random
 
first_name = "Jaya"
last_name = "Bodegard" 
 
def print_variables():
    random_number = random.randint(0,9)
    print(first_name)
    print(last_name)
    print(random_number)

print(globals())
    
````
